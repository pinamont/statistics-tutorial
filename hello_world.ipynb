And now something a bit more complicated, importing some ROOT and RooFit classes, filling and drawing a histogram, a RooFit plot and a TCanvas

```python
from ROOT import TCanvas, TH1F, TF1, RooFit, RooDataHist, RooRealVar

c = TCanvas("c","c",600,600)
x = RooRealVar("x","Observable",-10,10)
xframe = x.frame()

h = TH1F("h","Binned Data",200,-10,10)
g = TF1("g","gaus",-10,10)
g.SetParameters(1,0,2)
h.FillRandom("g",10000)
data = RooDataHist("data","Binned Data",x,RooFit.Import(h))
data.plotOn(xframe)
xframe.Draw()

c.Draw()
```
